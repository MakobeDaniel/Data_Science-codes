```python
df= data
df.head()
```

```python
df.select_dtypes('object').head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

df.NAME_CONTRACT_TYPE = le.fit_transform(df.NAME_CONTRACT_TYPE)
df.CODE_GENDER = le.fit_transform(df.CODE_GENDER)
df.FLAG_OWN_CAR = le.fit_transform(df.FLAG_OWN_CAR)
df.FLAG_OWN_REALTY = le.fit_transform(df.FLAG_OWN_REALTY)

df.NAME_TYPE_SUITE = le.fit_transform(df.NAME_TYPE_SUITE)
df.NAME_INCOME_TYPE = le.fit_transform(df.NAME_INCOME_TYPE)
df.NAME_EDUCATION_TYPE = le.fit_transform(df.NAME_EDUCATION_TYPE)
df.NAME_FAMILY_STATUS = le.fit_transform(df.NAME_FAMILY_STATUS)

df.NAME_HOUSING_TYPE = le.fit_transform(df.NAME_HOUSING_TYPE)
df.OCCUPATION_TYPE = le.fit_transform(df.OCCUPATION_TYPE)
df.WEEKDAY_APPR_PROCESS_START = le.fit_transform(df.WEEKDAY_APPR_PROCESS_START)
df.ORGANIZATION_TYPE = le.fit_transform(df.ORGANIZATION_TYPE)

df.FONDKAPREMONT_MODE = le.fit_transform(df.FONDKAPREMONT_MODE)
df.HOUSETYPE_MODE = le.fit_transform(df.HOUSETYPE_MODE)
df.WALLSMATERIAL_MODE = le.fit_transform(df.WALLSMATERIAL_MODE)
df.EMERGENCYSTATE_MODE = le.fit_transform(df.EMERGENCYSTATE_MODE)

```

```python
df.head()
```

```python
df.select_dtypes('object').head()
```

```python
df = df.replace(to_replace = np.nan, value = 0) 
#from sklearn.impute import SimpleImputer
#imp = SimpleImputer(missing_values=np.nan,strategy="mean")
#df = imp.fit_transform(df)
#df1 = pd.DataFrame(imp.fit_transform(df))
```

```python
df = df.drop("SK_ID_CURR",axis=1)
df
```

```python
#scaling the data
from sklearn.preprocessing import MinMaxScaler
# Scalling columns with values above 1 only.
scale = [col for col in df.columns if df[col].max() > 1]
MMS = MinMaxScaler()
scaled_df = MMS.fit_transform(df[scale])
scaled_df = pd.DataFrame(scaled_df, columns=scale)
## Replacing the original columns with the scaled columns
for col in scaled_df:
    df[col] = scaled_df[col]
    
df.head()
```

```python
print("The is now ready to be balanced")
```

```python
#BALANCING THE DATA
## FIRST SPLIT THE DATA SO THAT THERE IS NO BIAS IN THE TEST.

x=df.drop("TARGET",axis=1)
y=df.TARGET
#split data into training and testing sets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=42,test_size=.25)

```

```python
from imblearn.over_sampling import SMOTE 
SM = SMOTE(random_state=42)
X_SM, y_SM = SM.fit_resample(x_train, y_train)
```

```python
print("After implementing SMOTE")
print("The balanced data plotted")
y_SM.value_counts().plot.bar()
```

```python
X_SM.shape,y_SM.shape
```

```python
# train models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, recall_score, confusion_matrix
```

```python
# logistic regression
LR = LogisticRegression()

# RandomForest
RFC = RandomForestClassifier(random_state=42)
# fit model
model1= LR.fit(X_SM, y_SM)
model2= RFC.fit(X_SM, y_SM)
```

```python
x_test, y_test = SM.fit_resample(x_test, y_test)
y_pred1 = model1.predict(x_test)
y_pred2 = model2.predict(x_test)
print(y_pred1)
print(y_pred2)
```

```python
#Confussion matrix for sensitivity
print("Sensitivity of Model1")
print(f"Accuracy = {accuracy_score(y_test, y_pred1):.2f}\nRecall = {recall_score(y_test, y_pred1):.2f}\n")
CM = confusion_matrix(y_test, y_pred1)
plt.figure(figsize=(10, 6))
plt.title("Confusion Matrix", size=16)

sns.heatmap(CM, annot=True, cmap="Blues");
```

```python
#Confussion matrix for sensitivity
print("Sensitivity of Model2")
print(f"Accuracy2 = {accuracy_score(y_test, y_pred2):.2f}\nRecall = {recall_score(y_test, y_pred2):.2f}\n")
CM = confusion_matrix(y_test, y_pred2)
plt.figure(figsize=(10, 6))
plt.title("Confusion Matrix", size=16)

sns.heatmap(CM, annot=True, cmap="Blues");
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve

#predicting the probability
pred_prob1 = model1.predict_proba(x_test)
pred_prob2 = model2.predict_proba(x_test)

# roc curve for model
fpr1, tpr1, thresh1 = roc_curve(y_test, pred_prob1[:,1], pos_label=1)
fpr2, tpr2, thresh2 = roc_curve(y_test, pred_prob2[:,1], pos_label=1)

# roc curve for tpr = fpr 
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```

```python
auc_score1 = roc_auc_score(y_test, pred_prob1[:,1])
auc_score2 = roc_auc_score(y_test, pred_prob2[:,1])


print(auc_score1 ,auc_score2)
```

```python
# plotting roc curves
plt.style.use('seaborn')
plt.plot(fpr1, tpr1, linestyle='--',color='red', label='Logistic Regression')
plt.plot(fpr2, tpr2, linestyle='--',color='green', label='RFC')
plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
plt.title('ROC curve')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive rate")
plt.legend(loc="best")
plt.savefig("ROC",dpi=300)
plt.show();
```
